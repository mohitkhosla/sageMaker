```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
bostonHouseFrame=pd.read_csv('s3://rmit-ml-2020s1-lab-data/BostonHousingPrice/housing.data.csv',delimiter='\s+')

# bostonHouseFrame = pd.read_csv('housing.data.csv', delimiter='\s+')

bostonHouseFrame
```

## Create X & Y Variables

```python
X = bostonHouseFrame [['RM']]
Y = bostonHouseFrame[['MEDV']]
```

## Create testing and training data for RM variable

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(np.array(X), np.array(Y), test_size=0.2)
print(trainX.shape)
print(testX.shape)
print(trainY.shape)
print(testY.shape)
```

## Linear Regression Model

```python
from sklearn import linear_model
linReg = linear_model.LinearRegression()
```

```python
linReg.fit(trainX, trainY)
```

linReg.fit()fits the X and Y training data, and optimises the parameters to minimise the loss function. It might not exactly use gradient descent, but the ideas are similar and as stated in lectures, gradient descent as a general optimisation is the crux of many optimisation and parameter fitting algorithms.

## Parameters of Linear Regression

```python
print(linReg.intercept_)
print(linReg.coef_)
```

linReg.intercept_ is the y-intercept, or essentially the θ0 parameter. linReg.coef_ is slope of the univariate linear regression line, or the θ1 variable.

```python
predYRm = linReg.predict(testX)
```

predicts the Y value for each input X value intestXtesting dataset. The predictions are stored inpredYRm.

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error', mean_squared_error(testY, predYRm))
```

What we want to do is compare the predicted value from our model (given testX) with the actual Y values (testY). We can estimate the error (using themean squared error loss function for fitting theparameters)

```python
plt.rcParams['figure.figsize'] = [20, 10]
plt.figure()
plt.subplot(1,2,1)
plt.scatter(trainX, trainY)
plt.plot(trainX,linReg.predict(trainX),color='red',linewidth=7.0)
plt.xlabel("RM")
plt.ylabel("MDEV")
plt.title('Train Data')

plt.subplot(1,2,2)
plt.scatter(testX, testY)
plt.plot(testX,testY,color='red',linewidth=7.0)
plt.xlabel("RM")
plt.ylabel("MDEV")
plt.title('Test Data')

plt.show()
```

This uses matplotlib and pyplot to produce a SCATTERPLOT of our training/testing data (X,Y) pairs, and draw a red line for our model/hypothesis.

```python
plt.scatter(testX, linReg.predict(testX) - testY, c='g', s=40)
plt.hlines(y=0, xmin=3.5, xmax=9)
```

If the model is perfect, then for each testing data then there is 0 residual (or 0 error between predicted versus actual). This would produce a the horizontal line in the diagram. However, if the model under-estimates the
actual value, it will be below this horizontal line, and if over-estimates, then above. This plot can quickly giveyou a sense of where the errors are occurring and whether there are outlier points that are causing large errors.
